```python
#import necessary libraries/packages
from bs4 import BeautifulSoup
import requests
import pandas as pd
import sqlite3
```

```python
#1. Scrape all Pokemon data
url = 'https://pokemondb.net/pokedex/all'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
#this is the whole html code from the url
#print(soup)
```

```python
#find the table from the html code
table = soup.find('table')

#print(table)
```

```python
#Extract the Headers of the table
header_titles = table.find_all('th')
header_titles
```

```python
#strip unnecessary characters
header_table_titles = [title.text for title in header_titles]
#print(header_table_titles)
```

```python
df = pd.DataFrame(columns = header_table_titles)
df
```

```python
#find the all tr from tbody
tbody = table.find('tbody')
column_data = tbody.find_all('tr')
#print(column_data)
```

```python
#loop through each row
for row in column_data:
  #find all td's in a tr
  row_data = row.find_all('td')
  #extract text content from each td and strip unnecessary characters
  individual_row_data = [data.text.strip() for data in row_data]

  #print(individual_row_data)
  length = len(df)
  df.loc[length] = individual_row_data
```

```python
#2. Store the results in an SQL database.
# SQLite connection
conn = sqlite3.connect('pokedex.db')

# Insert DataFrame into SQLite
df.to_sql('pokemon_data', conn, if_exists='replace', index=False)

# Close connection
conn.close()

print("DataFrame successfully saved as db :).")
```
